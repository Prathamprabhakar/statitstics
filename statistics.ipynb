Statistics Basics

Conceptual Questions

1. What is statistics, and why is it important?
Statistics is the science of collecting, analyzing, interpreting, and presenting data. It is important for making informed decisions in fields such as business, healthcare, economics, and social sciences.


2. What are the two main types of statistics?
Descriptive statistics and inferential statistics.


3. What are descriptive statistics?
They summarize and describe the features of a dataset (e.g., mean, median, mode, variance).


4. What is inferential statistics?
It involves making predictions or inferences about a population based on a sample of data.


5. What is sampling in statistics?
Sampling is the process of selecting a subset from a population to estimate characteristics of the whole group.


6. What are the different types of sampling methods?
Simple random, stratified, systematic, cluster, and convenience sampling.


7. What is the difference between random and non-random sampling?
Random sampling gives all members an equal chance of selection; non-random does not and may introduce bias.


8. Define and give examples of qualitative and quantitative data.

Qualitative: Categorical data (e.g., gender, color)

Quantitative: Numerical data (e.g., height, age)*



9. What are the different types of data in statistics?
Nominal, ordinal, interval, and ratio.


10. Explain nominal, ordinal, interval, and ratio levels of measurement.

Nominal: Categories (e.g., gender)

Ordinal: Ordered categories (e.g., satisfaction levels)

Interval: Ordered, equal intervals, no true zero (e.g., temperature in Celsius)

Ratio: Like interval, with a true zero (e.g., weight)

Measures of Central Tendency and Dispersion

11. What is the measure of central tendency?
It represents the center of a data set: mean, median, mode.


12. Define mean, median, and mode.

Mean: Average

Median: Middle value

Mode: Most frequent value



13. What is the significance of the measure of central tendency?
It helps summarize a dataset with a single representative value.


14. What is variance, and how is it calculated?
Variance measures data spread. It's the average squared deviation from the mean.


15. What is standard deviation, and why is it important?
Standard deviation is the square root of variance. It shows how much data deviates from the mean.


16. Define and explain the term range in statistics.
Range = Max value – Min value. It shows total spread.


17. What is the difference between variance and standard deviation?
Variance is in squared units; standard deviation is in original units.

Data Distribution and Relationships

18. What is skewness in a dataset?
It indicates the asymmetry of the distribution.


19. What does it mean if a dataset is positively or negatively skewed?

Positively skewed: Tail on the right (mean > median)

Negatively skewed: Tail on the left (mean < median)



20. Define and explain kurtosis.
Kurtosis measures the "tailedness" of a distribution—whether data are heavy- or light-tailed relative to normal.


21. What is the purpose of covariance?
To measure the directional relationship between two variables.


22. What does correlation measure in statistics?
The strength and direction of the linear relationship between two variables.


23. What is the difference between covariance and correlation?
Correlation standardizes covariance, making it unit-free and easier to interpret.


24. What are some real-world applications of statistics?
Market research, quality control, medical trials, financial forecasting, social science studies.
